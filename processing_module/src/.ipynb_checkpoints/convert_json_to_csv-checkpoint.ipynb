# This notebook is used to convert Gold Standard bioc JSON file into CSV file with annotations

```python
import numpy as np
import pandas as pd
import json
```

```python
filename = "../results/supple.json"
f = open(filename, encoding="utf8")
data = json.load(f)
data
```

```python
for item in data["documents"]:
    print(item)
```

```python
filename = "../results/supple7.csv"
list_main = []
headers = ["pmid", "type", "id", "entity_type", "text", "location/reference", "infon"]
for elem in data["documents"]:
    
    for item in ((elem["passages"])):
        
        for annot in item["annotations"]:
            
            
            pmid_ = elem["id"]
            type_ = "annotation"
            id_ = annot["id"]
            entity_ = annot["infons"]["type"]
            text = annot["text"]
            offset_ = annot["locations"][0]["offset"]
            length_ = annot["locations"][0]["length"]
            loc_ref = "{}:{}".format(offset_, length_)
            
            infon = ""
            if "NCBI:txid" in annot["infons"]:
                infon = "NCBI:txid:{}".format(str(annot["infons"]["NCBI:txid"]))

            
            list_main.append([pmid_,type_,id_, entity_, text,loc_ref,infon])

df = pd.DataFrame(list_main, columns=headers)
df
            
```

```python
df.to_csv("../results/supplementary_file6.csv", index=None)
```
