```python
document_ids = set(df["Document"].to_list())
document_ids
```

```python
for docid in sorted(document_ids):
    print(str(docid) + "\n")
    df_temp = df[df["Document"]== docid]
#     print(df_temp)
    df_temp2 = df_temp[["Document", "Type", "Text"]]
#     print(df_temp2)
    print(df_temp2["Type"].value_counts())
    print(sum(df_temp2["Type"].value_counts()))
    
    print("\n")
```
