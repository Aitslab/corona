# This notebook is used to create a variants dictionary for covid 19 variants

```python
import numpy as np
import pandas as pd
import nltk 
```

```python
file1 = "../data/cov_variants/Cov-Lineages.html"
```

```python
df1 = pd.read_html(file1, header=0)[0]
```

```python
df1
```

```python
file2 = "../data/cov_variants/ncov_clades.tsv at master · nextstrain_ncov.htm"
```

```python
df2 = pd.read_html(file2)[0]
df2
```

```python
set1 = set(df1["Lineage"].tolist())
set2 = set(df2["clade"].tolist())
```

```python
set1
```

```python
set2
```

```python
set3 = set(["S", "L", "V", "G", "GK", "GH","GR", "GV","GRY"])
set3
```

```python
set_main = set.union(set1, set2,set3)
set_main
```

```python
len(set1), len(set2), len(set_main)
```

```python
filename = "../results/variants_all_processed.txt"
with open(filename, "w") as f:
    for term in sorted(set_main):
        f.write(term + "\n")
```
