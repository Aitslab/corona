```
import pandas as pd 
```

```
meta_df = pd.read_csv("metadata.csv", dtype={
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str, 
    'doi': str
})
meta_df.head()
```

```
from spacy.lang.en import English
nlp = English()
doc = nlp(meta_df.iloc[1]["abstract"])
```

```
meta_df.shape[0]
```

```
abstcts = []
for i in range(meta_df.shape[0]):
  t = meta_df.iloc[i]["abstract"]
  if t != t: #if there is no abstract pass
    pass
  else:  
    abstcts.append(t) #if there is indeed an abstract join the list
  print(i)
```

```
abstcts
```

```
from spacy.matcher import Matcher, PhraseMatcher
matcher = Matcher(nlp.vocab)
matcher.add("household",None,[{"LOWER": "household"}])
matcher.add("home",None,[{"LOWER": "home"}, {"LOWER": "care"}])
#matcher.add("care",None,[{"TEXT": "primary care"}, {'IS_ASCII': True, 'OP': '*'}]), 'household', {'IS_ASCII': True, 'OP': '*'}
#patt = [nlp(text) for text in ('household')]
#matcher.add('Home', None, *patt)
#matcher.add("community",None,[{"LOWER": "community"}])
#matcher.add("community",None,[{"LOWER": {'REGEX' : '/^(?!hospital)/'}}])
#doc = nlp(abstcts)
```

```
# the first filter for community/household category
m = []
for i in range(len(abstcts)):
  #print(i)
  doc = nlp(abstcts[i])
  matches = matcher(doc)
  if matches != []:
    print(i)
    m.append(doc)
```

```
# second filter for the community/household category
matcher = Matcher(nlp.vocab)
matcher.add("severity",None,[{"LOWER": "severity"}, {'IS_ASCII': True, 'OP': '*'}])
matcher.add("symptoms",None,[{"LOWER": "symptoms"}, {'IS_ASCII': True, 'OP': '*'}])
matcher.add("transmission",None,[{"LOWER": "transmission"}, {'IS_ASCII': True, 'OP': '*'}])
o = []
for i in range(len(m)):
  #print(i)
  doc = nlp(str(m[i]))
  matches = matcher(doc)
  if matches != []:
    print(i)
    o.append(doc)
```
